```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
product_cat = pd.read_csv('상품분류-D.txt', encoding='EUC-KR')
product_cat = product_cat.drop(['대분류코드', '중분류코드'], axis=1)
product_cat.head()
```

```python
purchase_tr = pd.read_csv('구매상품TR-D.csv', encoding='EUC-KR')
purchase_tr = purchase_tr.drop(['대분류코드', '중분류코드'], axis=1)
purchase_tr.head()
```

## 1. 데이터셋 재구성

```python
purchase_df = pd.merge(purchase_tr, product_cat, how='left', on='소분류코드')
purchase_df = purchase_df.drop('제휴사_y', axis=1).rename(columns={'제휴사_x':'제휴사'})
purchase_df.head()
```

```python
purchase_df['value'] = 1
purchase_df.head()
```

```python
purchase_pivot = pd.pivot_table(purchase_df, index='영수증번호', columns=['소분류명'], values=['value'], fill_value=0)
purchase_pivot
```

```python
purchase_pivot.columns = purchase_pivot.columns.droplevel(level=0)
purchase_pivot.head()
```

```python
purchase_pivot.columns
```

```python
pd.options.display.max_columns = 1000
purchase_pivot = purchase_pivot.reset_index()
purchase_pivot
```

```python
products_list = list(purchase_pivot.columns[1:])
products_list
```
