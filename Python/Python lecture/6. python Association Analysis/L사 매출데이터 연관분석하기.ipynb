```python
import numpy as np
import pandas as pd
product_cat = pd.read_csv("상품분류-D.txt", encoding="EUC-KR")
purchase_tr = pd.read_csv("구매상품TR-D.csv", encoding="euc-kr")
```

```python
purchase_tr.head()
```

```python
purchase_df = purchase_tr.merge(product_cat, how="left")
```

```python
purchase_df.head()
```

```python
purchase_df["value"] = 1
```

```python
purchase_df.head()
```

```python
pd.options.display.max_columns = 1000
```

```python
purchase_pivot = purchase_df.pivot_table(index="영수증번호", columns=["소분류명"], values=["value"], 
                                        fill_value=0)
purchase_pivot.head()
```

```python
열이름 = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
값 = [0,0,0,1,0,1,1]
```

```python
# 값이 1일 경우의 열 이름을 출력하세요.
result = []
for i, value in enumerate(값):
    if value==1:
        result.append(열이름[i])
```

```python
temp = purchase_pivot.iloc[4,:]
print(temp)
```

```python
purchase_pivot.columns = purchase_pivot.columns.droplevel(level=0)
```

```python
purchase_pivot.head()
```

```python
items = purchase_pivot.columns.values
```

```python
for i, value in enumerate(temp):
    if value==1:
        print(items[i])
```

```python
def get_item_name(row):
    item_names = []
    for i, value in enumerate(row):
        if value==1:
            item_names.append(items[i])
    return item_names
```

```python
def get_col_names(row):
    return [items[i] for i, value in enumerate(row) if value==1]
```

```python
transactions = purchase_pivot.apply(get_item_name, axis=1)
```

```python
purchase_pivot.apply(get_col_names, axis=1)
```

```python
from apyori import apriori
rules = apriori(transactions, min_support=0.001)
results = list(rules)
```

```python
len(results)
```

```python
result_df = pd.DataFrame(None, columns=["lhs", "rhs", "support", "confidence", "lift"])
```

```python
index = 0

for row in results:
    support = row[1]
    ordered_stat = row[2]
    for ordered_item in ordered_stat:
        lhs = " ".join(x.strip() for x in ordered_item[0])
        rhs = " ".join(x.strip() for x in ordered_item[1])
        confidence = ordered_item[2]
        lift = ordered_item[3]
        result_df.loc[index] = [lhs, rhs, support, confidence, lift]
        index += 1
        
```

```python
result_df.sort_values(by="lift", ascending=False)
```
