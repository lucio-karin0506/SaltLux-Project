```python
from matplotlib import rc
import matplotlib.pyplot as plt

rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
import numpy as np
import math

x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title("사인 웨이브")
plt.show()
```
