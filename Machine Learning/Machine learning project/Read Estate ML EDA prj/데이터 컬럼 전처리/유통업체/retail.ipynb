```python
import pandas as pd
```

```python
train_data = pd.read_csv('C:/Projects/python-workspace/Machine Learning/Machine learning project/team4 eda prj/dataset/train_data_ver1.csv')
train_data.head()
```

```python
retail_df = pd.read_csv('C:/Projects/python-workspace/Machine Learning/Machine learning project/team4 eda prj/dataset/유통업체.csv', encoding='cp949')
retail_df.head()
```

### 1. 구별 집계

```python
group_by_gu = retail_df.groupby('자치구').sum()
group_by_gu = group_by_gu.reset_index()
group_by_gu['전체 마트 수'] = group_by_gu['대형마트 수'] + group_by_gu['백화점 수']
group_by_gu = group_by_gu[group_by_gu['전체 마트 수'] != 0]
group_by_gu = group_by_gu.reset_index().drop('index', axis=1)
group_by_gu
```

```python
# 구별 대형마트 수 원본 데이터에 매핑
big_mart_count_by_gu = group_by_gu[['자치구', '대형마트 수']]
big_mart_guname = list(big_mart_count_by_gu['자치구'])
big_mart_guvalue = list(big_mart_count_by_gu['대형마트 수'])
big_mart_count_by_gu = pd.Series(index=big_mart_guname, data=big_mart_guvalue)

train_data['구별 대형마트 수'] = train_data['구'].map(big_mart_count_by_gu)
train_data['구별 대형마트 수'].value_counts()
```

```python
# 구별 백화점 수 원본 데이터에 매핑
department_count_by_gu = group_by_gu[['자치구', '백화점 수']]
department_guname = list(department_count_by_gu['자치구'])
department_guvalue = list(department_count_by_gu['백화점 수'])
department_count_by_gu = pd.Series(index=department_guname, data=department_guvalue)

train_data['구별 백화점 수'] = train_data['구'].map(department_count_by_gu)
train_data['구별 백화점 수'].value_counts()
```

```python
# 구별 전체 마트 수 원본 데이터에 매핑
all_count_by_gu = group_by_gu[['자치구', '전체 마트 수']]
all_guname = list(all_count_by_gu['자치구'])
all_guvalue = list(all_count_by_gu['전체 마트 수'])
all_count_by_gu = pd.Series(index=all_guname, data=all_guvalue)

train_data['구별 전체 마트 수'] = train_data['구'].map(all_count_by_gu)
train_data['구별 전체 마트 수'].value_counts()
```

### 2. 동별 집계

```python
group_by_dong = retail_df.groupby('동').sum()
group_by_dong = group_by_dong.reset_index()
group_by_dong['전체 마트 수'] = group_by_dong['대형마트 수'] + group_by_dong['백화점 수']
group_by_dong = group_by_dong[group_by_dong['전체 마트 수'] != 0]
group_by_dong = group_by_dong.reset_index().drop('index', axis=1)
group_by_dong
```

```python
# 동별 대형마트 수 원본 데이터에 매핑
big_mart_count_by_dong = group_by_dong[['동', '대형마트 수']]
big_mart_name = list(big_mart_count_by_dong['동'])
big_mart_value = list(big_mart_count_by_dong['대형마트 수'])
big_mart_count_by_dong = pd.Series(index=big_mart_name, data=big_mart_value)

big_mart_count_by_dong
```

```python
train_data['동별 대형마트 수'] = train_data['동'].map(big_mart_count_by_dong)
train_data['동별 대형마트 수'].value_counts()
```

```python
# 동별 백화점 수 원본 데이터에 매핑
department_count_by_dong = group_by_dong[['동', '백화점 수']]
department_name = list(department_count_by_dong['동'])
department_value = list(department_count_by_dong['백화점 수'])
department_count_by_dong = pd.Series(index=department_name, data=department_value)

department_count_by_dong
```

```python
train_data['동별 백화점 수'] = train_data['동'].map(department_count_by_dong)
train_data['동별 백화점 수'].value_counts()
```

```python
# 동별 전체 마트 수 원본 데이터에 매핑
all_count_by_dong = group_by_dong[['동', '전체 마트 수']]
all_name = list(all_count_by_dong['동'])
all_value = list(all_count_by_dong['전체 마트 수'])
all_count_by_dong = pd.Series(index=all_name, data=all_value)

all_count_by_dong
```

```python
train_data['동별 전체 마트 수'] = train_data['동'].map(all_count_by_dong)
train_data['동별 전체 마트 수'].value_counts()
```
